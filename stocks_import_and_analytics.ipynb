First we import data from Yahoo Finances

```python
all_data = ['NFLX','AMZN','BABA','AAPL','TWTR','SPY','BTC-USD']
```

```python
import pandas_datareader as data
import pandas as pd
```

```python
data_source = 'yahoo'
start_date = '2015-01-01'
end_date = '2022-08-01'
Netflix = data.DataReader('NFLX', data_source, start_date, end_date)
Apple = data.DataReader('AAPL', data_source, start_date, end_date)
AliBaba = data.DataReader('BABA', data_source, start_date, end_date)
Amazon = data.DataReader('AMZN', data_source, start_date, end_date)
Twitter = data.DataReader('TWTR', data_source, start_date, end_date)
Spy = data.DataReader('SPY', data_source, start_date, end_date)
Bitcoin = data.DataReader('BTC-USD', data_source, start_date, end_date)
```

```python
df = pd.concat([Netflix, Apple, AliBaba, Amazon, Twitter, Spy, Bitcoin], axis=1, keys=all_data)
df.head(5)
```

```python
#We will divide btc with 100 so it fits with other variableas.
df['BTC-USD'] = df['BTC-USD']/100
df['BTC-USD']
```

```python
#We will rename it so it's clear what its value is
df = df.rename(columns={'BTC-USD':'BTC-USD*100'})
```
